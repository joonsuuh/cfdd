```python
# libraries/packages
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl # for dracula color cycle
import scienceplots

plt.style.use(['science', 'no-latex'])
colorcycle = ['#8be9fd', '#ff79c6', '#50fa7b', '#bd93f9', '#ffb86c', '#ff5555', '#f1fa8c', 
'#6272a4']
dred = colorcycle[-3]
dgrey = colorcycle[-1]
plt.rcParams['axes.prop_cycle'] = mpl.cycler(color=colorcycle)

# change dpi 
plt.rcParams['figure.dpi'] = 200 # pixel density
plt.rcParams['savefig.dpi'] = 1024 # optional
```

```python
# first instability
relpath = './data/kelvin_helmholtz/'
# # load 2d data kh_output_vy.csv
# data = np.genfromtxt(relpath + 'kh_output_vy.csv', delimiter=',', filling_values=0)
# # max value of vy at each line 
# data = np.max(np.abs(data), axis=1)

data = np.genfromtxt(relpath + 'kh_output_vy.csv')

# # time every 0.01s
time = np.linspace(0, 2, len(data))

# vy propto e^sigma t, sigma = (2sqrt(1 + delta)) / (2 + delta) k v0
delta = 3
k = 6 * np.pi
v0 = 0.5
sigma = (2 * np.sqrt(1 + delta)) / (2 + delta) * k * v0
print(sigma)
vy_theory = np.exp(sigma * time) * 2e-3 # arbitrary constant since

# # plot
plt.plot(time, data, color=dred, label='Max $|v_y|$ along discontinuity')
plt.xlabel('$t$')
plt.ylabel('$|v_y|$')
# plt.title('First instability')
plt.yscale('log')
plt.plot(time, vy_theory, color=dgrey, linestyle='--', label='Theoretical growth')
# legend with box
plt.legend(loc='upper left', frameon=True)
plt.title('$k = 6\pi$, $v_0 = 0.5$, $\delta = 3$')
# plt.xlim(0, 1)
# plt.savefig('kh_growth_first.png')
# plt.savefig('kh_growth_k1.png')
# plt.savefig('kh_growth_k4.png')
# plt.savefig('kh_growth_k6.png')
# plt.savefig('kh_growth_k8.png')
# plt.savefig('kh_growth_k12.png')
plt.show()
```
