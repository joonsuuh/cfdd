```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import os
```

```python
relpath = './data/kelvin_helmholtz/'
data = np.loadtxt(relpath + 'output_rho_199.csv', delimiter=',')

# print(data.shape)

# find max and min
max_val = np.max(data)
min_val = np.min(data)
print(max_val, min_val)
# normalize data from -1 to 1
# data = (data - min_val) / (max_val - min_val) * 2.0 - 1

plt.imshow(data, cmap='viridis', origin='lower', extent=[0, 1, 0, 1])
plt.colorbar()
```

```python
csvs = [entry.name for entry in os.scandir(relpath) if "csv" in entry.name]
csvs.sort(key=lambda x: int(''.join(filter(str.isdigit, x))))
print(csvs)
count = 0
dt = 0.01


for csv in csvs:
    data = np.loadtxt(f'{relpath}/{csv}', delimiter=',')
    t = count * dt

    
    # find max and min
    max_val = np.max(data)
    min_val = np.min(data)
    # normalize data from -1 to 1
    data = (data - min_val) / (max_val - min_val) * 2.0 - 1

    fig, ax = plt.subplots()
    ax.set_title(f"$t$ = {t:.2f}", loc='right')
    plt.imshow(data, cmap='viridis', origin='lower', extent=[0, 1, 0, 1])
    plt.colorbar()
    
    count += 1
    plt.savefig(relpath + f'{count}.png')
    plt.close()    


    #run this in terminal:
    #ffmpeg -framerate 10 -i %d.png -c:v libx264 -vf "pad=ceil(iw/2)*2:ceil(ih/2)*2" -pix_fmt yuv420p kh.mp4
```
