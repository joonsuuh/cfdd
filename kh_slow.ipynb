```python
import scipy as sp
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import os
# set serif font
# import scienceplots

# plt.style.use(['science', 'no-latex'])
mpl.rcParams['font.family'] = 'serif'
plt.rcParams['savefig.dpi'] = 400
```

```python
relpath = './data/kelvin_helmholtz/'
time = 50
data = np.loadtxt(relpath + f'output_rho_{time}.csv', delimiter=',')
ttime = time / 100
# print(data.shape)

# find max and min
max_val = np.max(data)
min_val = np.min(data)
print(max_val, min_val)
# normalize data from -1 to 1
data = (data - min_val) / (max_val - min_val) * 2.0 - 1

fig, ax = plt.subplots()
plt.imshow(data, cmap='viridis', origin='lower', extent=[0, 1, 0, 1])
plt.colorbar()
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_title(r'$\rho$', loc='left')
ax.set_title(f'$t$ = {ttime:.2f}', loc='center')

# plt.savefig(f'kh_{time}.png')
```

```python
csvs = [entry.name for entry in os.scandir(relpath) if "csv" in entry.name]
csvs.sort(key=lambda x: int(''.join(filter(str.isdigit, x))))
print(csvs)
count = 0
dt = 0.01

for csv in csvs:
    data = np.loadtxt(f'{relpath}/{csv}', delimiter=',')
    t = count * dt

    
    # find max and min
    max_val = np.max(data)
    min_val = np.min(data)
    # normalize data from -1 to 1
    data = (data - min_val) / (max_val - min_val) * 2.0 - 1

    fig, ax = plt.subplots()
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_title(r'$\rho$', loc='left')
    ax.set_title(f'$t$ = {t:.2f}', loc='center')
    # ax.set_title(f"$t$ = {t:.2f}", loc='right')
    plt.imshow(data, cmap='viridis', origin='lower', extent=[0, 1, 0, 1])
    plt.colorbar()

    count += 1
    plt.savefig(relpath + f'{count:03d}.png')
    plt.close()    


# run this in terminal:
# ffmpeg -framerate 30 -i ./data/kelvin_helmholtz/%03d.png -c:v libx264 -vf "pad=ceil(iw/2)*2:ceil(ih/2)*2" -pix_fmt yuv420p kh_slow.mp4
```
